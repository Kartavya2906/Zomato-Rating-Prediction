# Inference-

## 1. Online Ordering & Rating Distribution
For the dominant rating range (2.5â€“5.0), the rating distribution follows an approximately normal pattern for both restaurants that support online ordering and those that do not. This indicates that while online ordering influences customer reach, the overall rating behavior remains statistically stable across both categories.

## 2. Impact of Location on Ratings
Ratings vary significantly across different locations, highlighting location as a strong explanatory feature. This observation justifies to use location-based encoding and supports its inclusion as a key categorical predictor in the model.

## 3. Table Booking & Customer Perception
In the 2.5â€“5.0 rating range, restaurants without table booking show a normal distribution of ratings, whereas restaurants offering table booking exhibit a higher concentration of ratings around 4.0. This suggests that table booking availability is positively associated with customer satisfaction.

## 4. Restaurant Type Influence
Ratings differ notably across restaurant types, confirming that rest_type captures meaningful variation in customer preferences. Selecting the dominant restaurant type per outlet and encoding it will be an effective feature engineering strategy.

## 5. Votes vs Rating Relationship
Higher-rated restaurants (ratings between 4.0 and 4.9) show an exponential increase in vote counts compared to lower-rated segments. This indicates strong user engagement and validates votes as an important numerical feature for predicting ratings.

## 6. Cuisine Diversity as a Key Driver
Cuisine composition has a substantial impact on ratings. Engineered features such as cuisines_count and popular_cuisines_count can effectively capture this influence while avoiding high-dimensional sparsity, making cuisines a critical predictor.

## 7. Cost vs Rating Trend
On average, higher-rated restaurants tend to have a higher approximate cost for two people. However, this trend is not consistent for lower-rated restaurants, where cost varies widely. This suggests a non-linear relationship between cost and rating rather than a direct correlation.

## 8. Listed City Consistency
Average ratings across different listed_in(city) categories fall within a narrow range (approximately 3.47â€“4.0), indicating minimal variation. This supports the decision to treat city listings as a lower-impact categorical feature but must be included for training as it impacts the Rating.

## 9. Listed Type Consistency
Similarly, listed_in(type) shows limited variation in average ratings (approximately 3.62â€“4.0), suggesting it provides contextual information but is not a dominant factor in rating prediction, although will be included by encoding in order to get more accurate Rating.
