# Modeling Implications from EDA

1. Numerical features such as votes, approx_cost, and cuisine counts were retained due to their strong association with ratings.

2. High-cardinality categorical variables were transformed using aggregation and dominant-category selection to reduce sparsity.

3. Location and restaurant type were encoded due to their demonstrated variability in ratings.

4. Overall, the EDA findings directly informed feature selection, transformation, and encoding strategies used to prepare the dataset for regression modeling
